```python
import matplotlib.pyplot as plt
import pickle


filename = "dbs_value_iteration_experiments_results.pkl"
with open(filename, "rb") as f:
    experiments_results = pickle.load(f)

COLOR_LIST = [ '#999999', '#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#000000' ]

for index, agent_name in enumerate(experiments_results['agent_name_list']):
    print("#{:d}: \'{:s}\'".format(index, agent_name))

first_list = range(8)
second_list = [ 4, 5, 6, 7, 8, 9 ]
static_list = [ 0, 1, 2, 3, 4 ]
dynamic_list = [ 7, 8, 9, 10, 11, 12 ]

# Static
fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for index, agent_index in enumerate(static_list):
    agent_name = experiments_results['agent_name_list'][agent_index]
    value_loss_array = experiments_results['value_loss_2darray'][agent_index]
    ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
ax.legend()
# fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
# for index, agent_index in enumerate(static_list):
#     agent_name = experiments_results['agent_name_list'][agent_index]
#     value_loss_array = experiments_results['value_loss_2darray'][agent_index]
#     ax.plot(range(1, 100 + 1), value_loss_array[:100], '.', markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
# ax.legend()

# Original (first list)
fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for index, agent_index in enumerate(first_list):
    agent_name = experiments_results['agent_name_list'][agent_index]
    value_loss_array = experiments_results['value_loss_2darray'][agent_index]
    ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
ax.legend()
# fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
# for index, agent_index in enumerate(first_list):
#     agent_name = experiments_results['agent_name_list'][agent_index]
#     value_loss_array = experiments_results['value_loss_2darray'][agent_index]
#     ax.plot(range(1, 100 + 1), value_loss_array[:100], '.', markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
# ax.legend()

# # (second list)
# fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
# for index, agent_index in enumerate(second_list):
#     agent_name = experiments_results['agent_name_list'][agent_index]
#     value_loss_array = experiments_results['value_loss_2darray'][agent_index]
#     ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
# ax.legend()
# fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
# for index, agent_index in enumerate(second_list):
#     agent_name = experiments_results['agent_name_list'][agent_index]
#     value_loss_array = experiments_results['value_loss_2darray'][agent_index]
#     ax.plot(range(1, 100 + 1), value_loss_array[:100], '.', markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
# ax.legend()

# Dynamic
fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for index, agent_index in enumerate(dynamic_list):
    agent_name = experiments_results['agent_name_list'][agent_index]
    value_loss_array = experiments_results['value_loss_2darray'][agent_index]
    ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
ax.legend()
fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for index, agent_index in enumerate(dynamic_list):
    agent_name = experiments_results['agent_name_list'][agent_index]
    value_loss_array = experiments_results['value_loss_2darray'][agent_index]
    ax.plot(range(1, 100 + 1), value_loss_array[:100], '.', markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
ax.legend()
fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for index, agent_index in enumerate(dynamic_list):
    agent_name = experiments_results['agent_name_list'][agent_index]
    value_loss_array = experiments_results['value_loss_2darray'][agent_index]
    ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
ax.legend()
ax.set_ylim(0.0, 1e-12)
```
