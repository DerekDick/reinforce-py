```python
import matplotlib.pyplot as plt
import pickle


filename = "dbs_value_iteration_value_loss_array.pkl"
with open(filename, "rb") as f:
    value_loss_array = pickle.load(f)


fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, '.', markevery=1000, linestyle='-', color='#00b5cc', label='DBS value iteration', markersize=12, linewidth=2)

# ax.plot(range(1, 1000 + 1), value_loss_array[:1000], '.', markevery=10, linestyle='-', color='blue', label='DBS value iteration', markersize=12, linewidth=2)

fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
ax.plot(range(1, 100 + 1), value_loss_array[:100], '.', markevery=1, linestyle='-', color='#00b5cc', label='DBS value iteration', markersize=12, linewidth=2)
```
