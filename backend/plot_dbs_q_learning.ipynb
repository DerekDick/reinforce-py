```python
import matplotlib.pyplot as plt
import pickle


filename = "dbs_q_learning_experiments_results.pkl"
with open(filename, "rb") as f:
    experiments_results = pickle.load(f)

COLOR_LIST = [ '#999999', '#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#000000' ]

fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for agent_index, agent_name in enumerate(experiments_results['agent_name_list']):
    averaged_number_of_step_to_goal_array = experiments_results['averaged_number_of_step_to_goal_2darray'][agent_index]
#     ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, '.', markevery=1, linestyle='-', color=COLOR_LIST[agent_index], label=agent_name, markersize=12, linewidth=2)
    ax.plot(range(1, averaged_number_of_step_to_goal_array.shape[0] + 1), averaged_number_of_step_to_goal_array, markevery=1, linestyle='-', color=COLOR_LIST[agent_index], label=agent_name, markersize=12, linewidth=2)
    ax.legend()

# ax.plot(range(1, 1000 + 1), value_loss_array[:1000], '.', markevery=10, linestyle='-', color='blue', label='DBS value iteration', markersize=12, linewidth=2)


fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
for agent_index, agent_name in enumerate(experiments_results['agent_name_list']):
    averaged_number_of_step_to_goal_array = experiments_results['averaged_number_of_step_to_goal_2darray'][agent_index]
    ax.plot(range(1, 100 + 1), averaged_number_of_step_to_goal_array[:100], '.', markevery=1, linestyle='-', color=COLOR_LIST[agent_index], label=agent_name, markersize=8, linewidth=2)
    ax.legend()
```
