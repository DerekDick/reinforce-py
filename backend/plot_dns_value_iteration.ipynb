```python
import matplotlib.pyplot as plt
import pickle

# mkdir
from pathlib import Path
Path("./figure").mkdir(parents=True, exist_ok=True)

filename = "dns_value_iteration_experiments_results.pkl"
with open(filename, "rb") as f:
    experiments_results = pickle.load(f)

COLOR_LIST = [ '#999999', '#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#000000' ]
MARKEVERY = 10
MARKER_SIZE = 25.0
LINE_WIDTH = 2.0
FONT_SIZE = 24.0
FONT_SIZE_TICK = 24.0
FIGSIZE = (16,9)
GRID = False
FONT_WEIGHT = "normal"

for index, agent_name in enumerate(experiments_results['agent_name_list']):
    print("#{:d}: \'{:s}\'".format(index, agent_name))

first_list = range(8)
# first_list = [ 0, 1, 2, 4, 5, 6 ]
# first_list = [ 0, 1, 2, 5, 6 ]

# All (first list)
fig, ax = plt.subplots(figsize=FIGSIZE, dpi=120)
for index, agent_index in enumerate(first_list):
    agent_name = experiments_results['agent_name_list'][agent_index]
    value_loss_array = experiments_results['value_loss_2darray'][agent_index]
    ax.plot(range(1, value_loss_array.shape[0] + 1), value_loss_array, markevery=MARKEVERY, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=MARKER_SIZE, linewidth=LINE_WIDTH)
ax.legend()
ax.set_xlabel(r'Step $t$', fontsize=FONT_SIZE, fontweight=FONT_WEIGHT)
ax.set_ylabel(r'Value Loss', fontsize=FONT_SIZE, fontweight=FONT_WEIGHT)
# Save the figure
fig.savefig('./figure/dns_value_iteration.pdf', format='pdf', bbox_inches='tight')
# fig, ax = plt.subplots(figsize=(16, 9), dpi=120)
# for index, agent_index in enumerate(static_list):
#     agent_name = experiments_results['agent_name_list'][agent_index]
#     value_loss_array = experiments_results['value_loss_2darray'][agent_index]
#     ax.plot(range(1, 100 + 1), value_loss_array[:100], '.', markevery=1, linestyle='-', color=COLOR_LIST[index], label=agent_name, markersize=12, linewidth=2)
# ax.legend()
```
